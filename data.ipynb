```python
import pandas as pd
```

```python
requests_file = "small.parquet"
requests_df = pd.read_parquet(requests_file)
symbol_idxs = [int(sym[-3:]) for sym in requests_df["symbol"].unique()]
num_symbols = max(symbol_idxs) + 1
print(f"{num_symbols} symbols")
```

```python
print(f"{len(requests_df)} rows")
requests_df.head()
```

```python
# Create large and small parquets
med_df = pd.concat([requests_df] * 10, ignore_index=True)
med_df = med_df.sample(frac=1).reset_index(drop=True)
med_df.to_parquet("med.parquet", index=False)

large_df = pd.concat([requests_df] * 100, ignore_index=True)
large_df = large_df.sample(frac=1).reset_index(drop=True)
large_df.to_parquet("large.parquet", index=False)

import os
os.path.exists("med.parquet"), os.path.exists("large.parquet")
```
